```python
# Merging DataFrames
"""
how : {'left', 'right', 'outer', 'inner', 'cross'}, default 'inner'
    Type of merge to be performed.

    * left: use only keys from left frame, similar to a SQL left outer join;
      preserve key order.
    * right: use only keys from right frame, similar to a SQL right outer join;
      preserve key order.
    * outer: use union of keys from both frames, similar to a SQL full outer
      join; sort keys lexicographically.
    * inner: use intersection of keys from both frames, similar to a SQL inner
      join; preserve the order of the left keys.
    * cross: creates the cartesian product from both frames, preserves the order
      of the left keys.
"""
```

```python
df1 = pd.DataFrame({
    'key1':[1,2,3,4,5],
     'key2':[6,7,8,9,10],
     'key3':[11,12,13,14,15]
})
df1
```

```python
df2 = pd.DataFrame({
    'key1':[1,23,33,4,53],
     'key4':[60,78,82,90,10],
     'key5':[11,122,103,14,105]
})
df2
```

```python
pd.merge(df1,df2, how = 'inner')
# Here common elements are 1 and 4 
# Notice, how the elements just next to them have been picked up
# in both dataframes
```

```python
pd.merge(df1,df2, how = 'left')
```

```python
pd.merge(df1,df2, how = 'right')
```

```python
pd.merge(df1,df2, how = 'outer')
# all the common elements of the common column from both the dataframes
```

```python
pd.merge(df1,df2, how = 'cross')
```

```python
pd.merge(df1,df2,how = 'left',left_on = "key2",right_on="key4")
```
