```python
# numerical data plotting
# relplot has options of plotting both categorical and numerical data
# relplot >> relationship b/w two variables
# relplot >> relationship plot
```

```python
# how tip is varying wrt total_bill >> scatter plot/ by default relplot makes scatter
```

```python
sbn.relplot(x = 'total_bill', y = 'tip' , data = df)
```

```python
sbn.relplot(x = 'total_bill', y = 'tip' , data = df, hue = 'smoker')
# insight >> as total bill increases bill increases
# customer who smoked gave the highest tip
```

```python
sbn.relplot(x = 'total_bill', y = 'tip' , data = df, hue = 'size')
# hue gives the different color sets
```

```python
sbn.relplot(x = 'total_bill', y = 'tip' , data = df, style = 'size')
```

```python
sbn.relplot(x = 'total_bill', y = 'tip' , data = df, style = 'smoker', hue = 'time')
```

```python
# line plot using relplot
time = np.arange(500)
value = np.random.randn(500).cumsum()
```

```python
df1 = pd.DataFrame({'time': time , 'value': value})
```

```python
df1
```

```python
sbn.relplot(x = 'time', y = 'value', kind = 'line', data = df1, sort = True )
```

```python
fmri = sbn.load_dataset('fmri')
```

```python
sbn.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri)
```

```python
sbn.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, ci = 'sd')
```

```python
sbn.relplot(x = 'timepoint', y = 'signal',estimator = None,kind = 'line', data = fmri, ci = 'sd')
```

```python
sbn.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, hue = 'event')
```

```python
sbn.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, hue = 'event', style = 'region')
```

```python

# dashes = False removes dashes in line
sbn.relplot(x = 'timepoint', y = 'signal', kind = 'line', dashes = False, data = fmri, hue = 'event', style = 'region')
```

```python

# dashes = False removes dashes in line
sbn.relplot(x = 'timepoint', y = 'signal', kind = 'line', dashes = False, data = fmri, hue = 'event',markers = True, style = 'region')
```

```python
tips = sbn.load_dataset('tips')
sbn.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = tips, col = 'time')
```

```python
sbn.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = tips, col = 'size')
```

```python
# for categorical data plots, we use catplot
```

```python
tips
```

```python
# plot between day and total_bill
sbn.catplot(x = 'day', y = 'total_bill', data = tips)
# insights >> many ppl arrve on weekends compared to week days
```

```python
# in catplot few datapts are overlapping, for clarity lets use swarm plot 
sbn.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm')
```

```python
# boxplot >> to see outliers
sbn.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'box', color = 'r')
```

```python
sbn.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'violin', color = 'm')
```

```python
# boxen plot is made from box plot
sbn.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'boxen', color = 'y')
```

```python
tips
```

```python
sbn.barplot(x = 'day', y = 'total_bill', data = tips, color = 'orange')
# toal bill aggregated on different days
```

```python
sbn.jointplot( x = tips.total_bill , y = tips.tip) # draw a plot of two variables
# with distribution >> bivariate plots
```

```python
# joint plot >> to understand relationship and also distribution of two varaibles
```

```python
sbn.jointplot(x = tips.total_bill , y = tips.tip, kind = 'reg')
```

```python
sbn.jointplot(x = tips.total_bill , y = tips.tip, kind = 'hex')
```

```python
sbn.jointplot(x = tips.total_bill , y = tips.tip, kind = 'kde')
# kernel density estimator, contour plot used in meteorological dept.
```

```python
sbn.pairplot(tips) # make scatter plots wrt all numerical colums
```

```python
sbn.pairplot(tips,hue = 'smoker')
```

```python
sbn.pairplot(tips,hue = 'smoker', palette = 'coolwarm' )
```

```python
# boxplot using sns directly(earlier catplot)
sbn.boxplot(x = 'day', y = 'total_bill', data = tips, palette = 'rainbow')
```

```python
sbn.boxplot(data = tips, palette = 'rainbow', orient = 'h')
```
